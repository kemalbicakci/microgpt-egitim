## 1. Kütüphaneler ve rastgelelik kontrolü

```
import os
import math
import random

# Deneyin tekrar üretilebilir olması için sabit seed
random.seed(42)
```

Bu hücrede:
- `os`: dosya kontrolü
- `math`: log, exp vb.
- `random`: veri karıştırma, örnekleme, başlangıç ağırlıkları

kullanılıyor.

`random.seed(42)` aynı kodun benzer sonuçlar vermesini kolaylaştırır.


## 2. Veri kümesini yükleme

```
# Eğer input.txt yoksa küçük isim veri kümesini indir
if not os.path.exists("input.txt"):
    import urllib.request
    names_url = "https://raw.githubusercontent.com/karpathy/makemore/988aa59/names.txt"
    urllib.request.urlretrieve(names_url, "input.txt")

docs = [line.strip() for line in open("input.txt") if line.strip()]
random.shuffle(docs)

print(f"num docs: {len(docs)}")
print("ilk 10 örnek:", docs[:10])
```

Burada modelin eğitileceği doküman listesi hazırlanıyor.

Bu örnekte her satır bir isimdir. Yani model aslında:
- karakter dizilerini görüyor
- bir sonraki karakteri tahmin etmeyi öğreniyor

Bu nedenle bu model bir **character-level language model** olarak düşünülebilir.


## 3. Tokenizer: karakterleri sayılara çevirme

```
# Veri kümesindeki tüm benzersiz karakterleri topla
uchars = sorted(set("".join(docs)))

# BOS = Beginning of Sequence için özel token
BOS = len(uchars)

# Toplam token sayısı
vocab_size = len(uchars) + 1

print("benzersiz karakterler:", uchars)
print("BOS token id:", BOS)
print("vocab size:", vocab_size)
```

Model metinle doğrudan çalışmaz; sayılarla çalışır.

Bu yüzden:
- her karaktere bir tamsayı kimliği verilir
- ayrıca özel bir `BOS` token'ı tanımlanır

Örnek mantık:
- `a -> 0`
- `b -> 1`
- ...
- `BOS -> son id`

Bu token, dizinin başladığını belirtmek için kullanılır.


## 4. Minimal autograd motoru: `Value` sınıfı

```
class Value:
    __slots__ = ("data", "grad", "_children", "_local_grads")

    def __init__(self, data, children=(), local_grads=()):
        self.data = data
        self.grad = 0
        self._children = children
        self._local_grads = local_grads

    def __add__(self, other):
        other = other if isinstance(other, Value) else Value(other)
        return Value(self.data + other.data, (self, other), (1, 1))

    def __mul__(self, other):
        other = other if isinstance(other, Value) else Value(other)
        return Value(self.data * other.data, (self, other), (other.data, self.data))

    def __pow__(self, other):
        return Value(self.data**other, (self,), (other * self.data**(other - 1),))

    def log(self):
        return Value(math.log(self.data), (self,), (1 / self.data,))

    def exp(self):
        return Value(math.exp(self.data), (self,), (math.exp(self.data),))

    def relu(self):
        return Value(max(0, self.data), (self,), (float(self.data > 0),))

    def __neg__(self): return self * -1
    def __radd__(self, other): return self + other
    def __sub__(self, other): return self + (-other)
    def __rsub__(self, other): return other + (-self)
    def __rmul__(self, other): return self * other
    def __truediv__(self, other): return self * other**-1
    def __rtruediv__(self, other): return other * self**-1

    def backward(self):
        topo = []
        visited = set()

        def build_topo(v):
            if v not in visited:
                visited.add(v)
                for child in v._children:
                    build_topo(child)
                topo.append(v)

        build_topo(self)
        self.grad = 1

        for v in reversed(topo):
            for child, local_grad in zip(v._children, v._local_grads):
                child.grad += local_grad * v.grad
```

Bu sınıf, çok basit bir **otomatik türev alma (autograd)** sistemi kurar.

Her `Value` nesnesi şunları taşır:
- `data`: ileri yayılımda oluşan sayı
- `grad`: geri yayılımda hesaplanan türev
- `_children`: bu değerin hangi önceki düğümlerden geldiği
- `_local_grads`: yerel türevler

### Neden önemli?
Modern derin öğrenmede geri yayılımın özü budur:
hesaplama grafiğini kur ve loss'tan başlayarak gradyanları geriye doğru dağıt.


## 5. Autograd için mini örnek

```
a = Value(2.0)
b = Value(3.0)
c = a * b + a
c.backward()

print("c.data =", c.data)
print("dc/da =", a.grad)  # c = a*b + a => dc/da = b + 1 = 4
print("dc/db =", b.grad)  # dc/db = a = 2
```

Bu mini örnek, `backward()` çağrısının gerçekten türev hesapladığını göstermektedir.


## 6. Model hiperparametreleri

```
n_layer = 1      # transformer layer sayısı
n_embd = 16        # embedding boyutu
block_size = 16    # maksimum bağlam uzunluğu
n_head = 4         # attention head sayısı
head_dim = n_embd // n_head

print("n_layer =", n_layer)
print("n_embd =", n_embd)
print("block_size =", block_size)
print("n_head =", n_head)
print("head_dim =", head_dim)
```

Bu değerler modeli çok küçük tutar.

Büyük LLM'lerde bunlar devasa olur; burada ise amaç eğitim ve açıklanabilirliktir.

- `n_embd`: her token'ın vektör boyutu
- `block_size`: modelin aynı anda görebildiği maksimum geçmiş uzunluğu
- `n_head`: multi-head attention içindeki başlık sayısı
- `head_dim`: her başlığa düşen boyut


## 7. Parametre matrisi oluşturucu

```
def matrix(nout, nin, std=0.08):
    return [[Value(random.gauss(0, std)) for _ in range(nin)] for _ in range(nout)]
```

Bu yardımcı fonksiyon, rastgele başlatılmış ağırlık matrisleri üretir.

Her eleman:
- Python float değil,
- `Value` nesnesidir

Böylece bütün hesaplama grafiği üzerinden gradyan taşınabilir.


## 8. Model ağırlıkları (`state_dict`)

```
state_dict = {
    "wte": matrix(vocab_size, n_embd),    # token embeddings
    "wpe": matrix(block_size, n_embd),    # position embeddings
    "lm_head": matrix(vocab_size, n_embd) # son projeksiyon
}

for i in range(n_layer):
    state_dict[f"layer{i}.attn_wq"] = matrix(n_embd, n_embd)
    state_dict[f"layer{i}.attn_wk"] = matrix(n_embd, n_embd)
    state_dict[f"layer{i}.attn_wv"] = matrix(n_embd, n_embd)
    state_dict[f"layer{i}.attn_wo"] = matrix(n_embd, n_embd)
    state_dict[f"layer{i}.mlp_fc1"] = matrix(4 * n_embd, n_embd)
    state_dict[f"layer{i}.mlp_fc2"] = matrix(n_embd, 4 * n_embd)

params = [p for mat in state_dict.values() for row in mat for p in row]
print(f"num params: {len(params)}")
print("state_dict anahtarları:", list(state_dict.keys()))
```

Burada modelin öğrenilecek tüm parametreleri oluşturuluyor.

Temel gruplar:
- `wte`: token embedding tablosu
- `wpe`: position embedding tablosu
- attention için: `Wq`, `Wk`, `Wv`, `Wo`
- MLP için: iki lineer katman
- `lm_head`: son gizli temsilden vocabulary logits üretir


## 9. Yardımcı fonksiyonlar: linear, softmax, rmsnorm

```
def linear(x, w):
    return [sum(wi * xi for wi, xi in zip(wo, x)) for wo in w]

def softmax(logits):
    max_val = max(val.data for val in logits)
    exps = [(val - max_val).exp() for val in logits]
    total = sum(exps)
    return [e / total for e in exps]

def rmsnorm(x):
    ms = sum(xi * xi for xi in x) / len(x)
    scale = (ms + 1e-5) ** -0.5
    return [xi * scale for xi in x]
```

### `linear`
Bir vektörü ağırlık matrisiyle çarpar.

### `softmax`
Logit'leri olasılıklara çevirir.
Matematiksel olarak:
\[
p_i = rac{e^{z_i}}{\sum_j e^{z_j}}
\]

### `rmsnorm`
LayerNorm benzeri ama daha sade bir normalizasyon yaklaşımıdır.
Bu implementasyonda GPT-2'deki LayerNorm yerine RMSNorm tercih edilmiştir.


## 10. GPT forward pass

```
def gpt(token_id, pos_id, keys, values):
    # Token ve pozisyon embeddinglerini al
    tok_emb = state_dict["wte"][token_id]
    pos_emb = state_dict["wpe"][pos_id]

    # Birleştir
    x = [t + p for t, p in zip(tok_emb, pos_emb)]
    x = rmsnorm(x)

    for li in range(n_layer):
        # ---- Multi-head attention bloğu ----
        x_residual = x
        x = rmsnorm(x)

        q = linear(x, state_dict[f"layer{li}.attn_wq"])
        k = linear(x, state_dict[f"layer{li}.attn_wk"])
        v = linear(x, state_dict[f"layer{li}.attn_wv"])

        keys[li].append(k)
        values[li].append(v)

        x_attn = []
        for h in range(n_head):
            hs = h * head_dim
            q_h = q[hs:hs + head_dim]
            k_h = [ki[hs:hs + head_dim] for ki in keys[li]]
            v_h = [vi[hs:hs + head_dim] for vi in values[li]]

            attn_logits = [
                sum(q_h[j] * k_h[t][j] for j in range(head_dim)) / head_dim**0.5
                for t in range(len(k_h))
            ]

            attn_weights = softmax(attn_logits)

            head_out = [
                sum(attn_weights[t] * v_h[t][j] for t in range(len(v_h)))
                for j in range(head_dim)
            ]
            x_attn.extend(head_out)

        x = linear(x_attn, state_dict[f"layer{li}.attn_wo"])
        x = [a + b for a, b in zip(x, x_residual)]

        # ---- MLP bloğu ----
        x_residual = x
        x = rmsnorm(x)
        x = linear(x, state_dict[f"layer{li}.mlp_fc1"])
        x = [xi.relu() for xi in x]
        x = linear(x, state_dict[f"layer{li}.mlp_fc2"])
        x = [a + b for a, b in zip(x, x_residual)]

    logits = linear(x, state_dict["lm_head"])
    return logits
```

Bu hücre notebook’un merkezidir.

Forward pass sırasında:
1. token embedding + position embedding alınır  
2. attention bloğundan geçirilir  
3. residual bağlantı eklenir  
4. MLP bloğundan geçirilir  
5. vocabulary logits üretilir  

### Attention sezgisi
Mevcut token, geçmişteki token’lara:
- neye bakacağını (`Q`)
- eldeki geçmiş temsilin ne olduğunu (`K`)
- hangi bilgiyi taşıyacağını (`V`)

üzerinden karar verir.

Skor yaklaşık olarak:
\[
\text{score}(q,k) = \frac{q \cdot k}{\sqrt{d}}
\]

Sonra softmax ile ağırlıklara çevrilir.


## 11. Tek adımlık forward örneği

```
example_doc = docs[0]
example_tokens = [BOS] + [uchars.index(ch) for ch in example_doc] + [BOS]

keys = [[] for _ in range(n_layer)]
values = [[] for _ in range(n_layer)]

token_id = example_tokens[0]
pos_id = 0

logits = gpt(token_id, pos_id, keys, values)

print("örnek doküman:", example_doc)
print("ilk token id:", token_id)
print("üretilen logit sayısı:", len(logits))
print("vocab_size ile aynı mı?", len(logits) == vocab_size)
```

Her ileri geçişte çıktı olarak `vocab_size` uzunluğunda bir logit vektörü elde edilir.
Bu vektör, sıradaki token için puanlardır.


## 12. Optimizer: Adam

```
learning_rate, beta1, beta2, eps_adam = 0.01, 0.85, 0.99, 1e-8

m = [0.0] * len(params)  # birinci moment
v = [0.0] * len(params)  # ikinci moment

print("optimizer hazır, parametre sayısı:", len(params))
```

Adam optimizer her parametre için geçmiş gradyan bilgisini tutar:

- `m`: ortalama gradyan (momentum benzeri)
- `v`: kareli gradyan ortalaması

Bu sayede güncellemeler daha kararlı olur.


## 13. Eğitim döngüsü

```
num_steps = 1000

for step in range(num_steps):
    # Tek bir doküman seç
    doc = docs[step % len(docs)]

    # BOS ile başlat ve bitir
    tokens = [BOS] + [uchars.index(ch) for ch in doc] + [BOS]

    # Kontekst sınırı
    n = min(block_size, len(tokens) - 1)

    # Her layer için cache
    keys = [[] for _ in range(n_layer)]
    values = [[] for _ in range(n_layer)]

    losses = []

    # Oto-regresif eğitim: sıradaki karakteri tahmin et
    for pos_id in range(n):
        token_id, target_id = tokens[pos_id], tokens[pos_id + 1]

        logits = gpt(token_id, pos_id, keys, values)
        probs = softmax(logits)

        # Cross-entropy'nin tek örnek hali: -log p(target)
        loss_t = -probs[target_id].log()
        losses.append(loss_t)

    loss = (1 / n) * sum(losses)

    # Backprop
    loss.backward()

    # Adam güncellemesi
    lr_t = learning_rate * (1 - step / num_steps)
    for i, p in enumerate(params):
        m[i] = beta1 * m[i] + (1 - beta1) * p.grad
        v[i] = beta2 * v[i] + (1 - beta2) * p.grad ** 2

        m_hat = m[i] / (1 - beta1 ** (step + 1))
        v_hat = v[i] / (1 - beta2 ** (step + 1))

        p.data -= lr_t * m_hat / (v_hat ** 0.5 + eps_adam)
        p.grad = 0

    if step % 50 == 0:
        print(f"step {step:4d} | loss {loss.data:.4f}")
```

Bu eğitim döngüsünde model şunu öğrenir:

> “Şu ana kadarki karakterleri gördüm; sıradaki karakter ne olabilir?”

Bu, klasik bir **causal / autoregressive language modeling** kurulumudur.


## 14. Loss mantığı

Tek zaman adımında kayıp şu şekildedir:

\[
L_t = -\log p(\text{hedef token})
\]

Tüm dizi için ortalama alınır:

\[
L = \frac{1}{n}\sum_{t=1}^{n} L_t
\]

Yani model doğru karaktere yüksek olasılık verirse loss küçülür.


## 15. İnference / örnek üretme

```
temperature = 0.5

print("\n--- inference (new, hallucinated names) ---")
for sample_idx in range(20):
    keys = [[] for _ in range(n_layer)]
    values = [[] for _ in range(n_layer)]

    token_id = BOS
    sample = []

    for pos_id in range(block_size):
        logits = gpt(token_id, pos_id, keys, values)
        probs = softmax([l / temperature for l in logits])

        token_id = random.choices(
            range(vocab_size),
            weights=[p.data for p in probs]
        )[0]

        if token_id == BOS:
            break

        sample.append(uchars[token_id])

    print(f"sample {sample_idx+1:2d}: {''.join(sample)}")
```

Inference aşamasında model artık öğrenmiyor; sadece yeni örnekler üretiyor.

### Temperature ne yapar?
- düşük temperature → daha tutucu, daha güvenli seçimler
- yüksek temperature → daha çeşitli, daha yaratıcı ama daha düzensiz çıktılar
